```python
#clean the text
!pip install nltk
from textblob import TextBlob
from wordcloud import WordCloud
import numpy as np
import nltk
import string
import re #regex library
#import word_tokenize & freqDist from NLTK
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import tweepy
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
nltk.download("punkt")


def remove_tweet_special(text):
  #remove tab, new line, ans back slice
  text = text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"")
  #remove non ASCII (Emoticon, Chinese word, dll)
  text = text.encode('ascii', 'replace').decode('ascii')
  #remove mention, link, hashtag
  text = ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)&amp"," ", text).split())
  #remove incomplete URL
  return text.replace("http://", " ").replace("https://", " ")

df['Tweets'] = df['Tweets'].apply(remove_tweet_special)

#remove number
def remove_number(text):
  return re.sub(r"\d+", "", text)

df['Tweets'] = df['Tweets'].apply(remove_number)


#remove punctuation
def remove_punctuation(text):
  return text.translate(str.maketrans("","",string.punctuation))

df['Tweets'] = df['Tweets'].apply(remove_punctuation)


#remove whitespace leading & trailing
def remove_whitespace_LT(text):
  return text.strip()

df['Tweets'] = df['Tweets'].apply(remove_whitespace_LT)


#remove multiple whitespace into single whitespace
def remove_whitespace_multiple(text):
  return re.sub('\s+',' ', text)

df['Tweets'] = df['Tweets'].apply(remove_whitespace_multiple)


#remove single char
def remove_single_char(text):
  return re.sub(r"\b[a-zA-Z]\b", "", text)

df['Tweets'] = df['Tweets'].apply(remove_single_char)


#create a function to clean the tweets
#def cleanTxt(text):
#  text = re.sub(r'@[A-Za-z0-9]+', '', text) #Removed @mention
#  text = re.sub(r'#', '', text) #Removing the '#' Symbol
#  text = re.sub(r'RT[\s]+', '', text) #Removing RT
#  text = re.sub(r'https?:\/\/\S+', '', text) #Removing the hyper link

#  return text

#cleaning the text
#df['Tweets']= df['Tweets'].apply(cleanTxt)
```

```python
import pandas as pd
import numpy as np

Tweet_TF = pd.read_csv("Hasil_Kami_Muak_Kaesang.csv", usecols = ["Tokens_Stemmed_Tweets"])
Tweet_TF.columns = ["Tweet"]


Tweet_TF.head()
```

```python
#convert list format string to list

import ast

def convert_text_list(texts):
  texts = ast.literal_eval(texts)
  return [text for text in texts]

Tweet_TF["Tweet_List"] = Tweet_TF["Tweet"].apply(convert_text_list)

print (Tweet_TF["Tweet_List"][90])
print("\ntype : ", type(Tweet_TF["Tweet_List"][90]))
#['potensi', 'bocor', 'rubah']
#type : <class 'list'>
```

```python
Tweet_TF.head()
```

```python
def calc_TF(document):
  #counts the number of times the word appears in review
  TF_dict = {}
  for term in document:
    if term in TF_dict:
      TF_dict[term] += 1
    else :
      TF_dict[term] = 1
  #compute TF for each word
  for term in TF_dict:
    TF_dict[term] = TF_dict[term] / len(document)
  return TF_dict

Tweet_TF["TF_dict"] = Tweet_TF['Tweet_List'].apply(calc_TF)
Tweet_TF["TF_dict"].head()
```

```python
#cek TF result
index = 90

print('%20s' % "term", "\t", "TF\n")
for key in Tweet_TF["TF_dict"][index]:
  print('%20s' % key, "\t", Tweet_TF["TF_dict"][index][key])
```

```python
def calc_DF(tfDict):
  count_DF = {}
  #berfungsi untuk mengurangi bobot data term jika kemunculan banyak tersebar
  #disuatu dataset
  for document in tfDict:
    for term in document:
      if term in count_DF:
        count_DF[term] += 1
      else :
        count_DF[term] = 1
  return count_DF

DF = calc_DF(Tweet_TF["TF_dict"])
print(DF)
```

```python
n_document = len(Tweet_TF)

def calc_IDF(__n_document, __DF):
  IDF_Dict = {}
  for term in __DF:
    IDF_Dict[term] = np.log(__n_document / (__DF[term] + 1))
  return IDF_Dict

#Stores the IDF dictionary
IDF = calc_IDF(n_document, DF)
```

```python
def calc_TF_IDF(TF):
    TF_IDF_Dict = {}

    #for each word in the review, we multiply its TF and its IDF
    for key in TF:
      TF_IDF_Dict[key] = TF[key] * IDF[key]
    return TF_IDF_Dict

    #stores the TF-IDF series
Tweet_TF["TF-IDF_dict"] = Tweet_TF["TF_dict"].apply(calc_TF_IDF)
```

```python
#CHECK TF-IDF result
index = 90

print('%20s' % "term", "\t", '%10s' % "TF", "\t", '%20s' % "TF-IDF\n")
for key in Tweet_TF["TF-IDF_dict"][index]:
   print('%20s' % key, "\t", Tweet_TF["TF_dict"][index][key],"\t" , Tweet_TF["TF-IDF_dict"][index][key])


#index = 90

#print('%20s' % "term", "\t", '%10S' % "TF", "\t", '%20s' % "TF-IDF\n")
#for key in Tweet_TF["TF_dict"][index]:
#  print('%20s' % key, "\t", Tweet_TF["TF_dict"][index][key], "\t", Tweet_TF["TF-IDF_dict"])
```

```python
#short desc by value for DF dictionary
sorted_DF = sorted(DF.items(), key = lambda kv: kv[1], reverse=True)[:50]

#create a list of unique words from sorted dictionary 'sorted_DF'
unique_term = [item[0] for item in sorted_DF]

def calc_TF_IDF_Vec(__TF_IDF_Dict):
  TF_IDF_vector = [0.0] * len (unique_term)

  #for each unique word, if it is in the review, store its TF-IDF value.
  for i, term in enumerate(unique_term):
    if term in __TF_IDF_Dict:
      TF_IDF_vector[i] = __TF_IDF_Dict[term]
  return TF_IDF_vector

Tweet_TF["TF_IDF_Vec"] = Tweet_TF["TF-IDF_dict"].apply(calc_TF_IDF_Vec)

print("print first row matrix TF_IDF_Vec Series\n")
print(Tweet_TF["TF_IDF_Vec"][0])

print("\nMatrix size : ", len (Tweet_TF["TF_IDF_Vec"][0]))
```

```python
#convert series to list
TF_IDF_Vec_List = np.array(Tweet_TF["TF_IDF_Vec"].to_list())

#sum element vector in axis=0
sums = TF_IDF_Vec_List.sum(axis=0)

data = []

for col, term in enumerate(unique_term):
  data.append((term, sums[col]))

ranking = pd.DataFrame(data, columns = ['term', 'rank'])
ranking.sort_values('rank', ascending=False)
```
